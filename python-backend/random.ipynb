```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Time': pd.date_range(start='1/1/2022', periods=300, freq='H'),
}

df = pd.DataFrame(data)

print(df)
```

```python
df['Water_Level'] = df.iloc[:, 1:].sum(axis=1)

print(df)
```

```python
df.columns.drop('Total')
```

```python
import matplotlib.pyplot as plt
df.ECE.plot()
df.CSE.plot()
df.EEE.plot()
df.Total.plot()
```
